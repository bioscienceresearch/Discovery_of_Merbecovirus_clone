```python
import pathlib
import pylab as plt
import os
import numpy as np
from numpy import unique
from numpy import where
import pandas as pd
import matplotlib
import seaborn as sns
import sys
```

```python
sns.set_style('white')
```

```python
df=pd.read_csv('data/Restrictionenzymes-Sheet1.csv')
```

```python
df = df.drop('Unnamed: 0', axis=1)
```

```python
bsaI=df.iloc[0].to_numpy() 
```

```python
bsmBI=df.iloc[1].to_numpy() 
```

```python
covs=list(df.columns.values)
```

```python
plt.rcParams["figure.figsize"] = (14,5)
```

```python

plt.bar(covs, bsaI, color='salmon', alpha=0.8, label='BsaI')
plt.bar(covs, bsmBI, bottom=bsaI, color='royalblue', alpha=0.88, label='BsmBI')
plt.xticks(rotation=90)
plt.legend(loc="upper left")
plt.savefig("plots/re_counts.png", bbox_inches='tight')
plt.show()
```
